# Hand Gesture Recognition


## Getting Hand Gestures from both the hands for Drone Simulation


```python
import cv2
import numpy as np
import datetime
import time
import pyautogui
import webbrowser
import matplotlib.pyplot as plt
```

```python
vid = cv2.VideoCapture(0)
l_r = np.array([0,48,80],dtype="uint8")
u_r = np.array([30,255,255],dtype="uint8")
kernel = (13,13)
while True:
    b,frame = vid.read()
    frame = cv2.flip(frame,1)
    
    #Control1
    img1 = frame[10:300,400:700]

    hsv1 = cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
    
    skin1 = cv2.inRange(hsv1,l_r,u_r)
   
    skin1 = cv2.erode(skin1,kernel,iterations=3)
    skin1 = cv2.dilate(skin1,kernel,iterations=5)

    skin1 = cv2.medianBlur(skin1,13)
    cont1,hier = cv2.findContours(skin1,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    if(len(cont1)!=0):
        cont1 = max(cont1,key=lambda x: cv2.contourArea(x))
        cv2.drawContours(img1,[cont1],-1,(255,255,0),2)

        hull1 = cv2.convexHull(cont1)
        cv2.drawContours(img1,[hull1],-1,(0,255,255),2)


        t1=tuple(hull1[hull1[:,:,1].argmin()][0])
        b1=tuple(hull1[hull1[:,:,1].argmax()][0])
        l1=tuple(hull1[hull1[:,:,0].argmin()][0])
        r1=tuple(hull1[hull1[:,:,0].argmax()][0])
        
        lr1 = np.linalg.norm(l1[-1]-r1[-1])
        lt1 = np.linalg.norm(l1[-1]-t1[-1])
        br1 = np.linalg.norm(b1[-1]-r1[-1])
        rt1 = np.linalg.norm(t1[-1]-r1[-1])
        bl1 = np.linalg.norm(l1[-1]-b1[-1])
        tb1 = np.linalg.norm(t1[-1]-b1[-1])
        
        
        if(tb1<150):
            cv2.putText(frame,"Forward",(400,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2,cv2.LINE_AA)
            pyautogui.keyDown('k')
        
            pyautogui.keyUp('k')
        
        elif(tb1>150 and lt1<90 and br1 < 90):
            
            cv2.putText(frame,"Left",(400,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2,cv2.LINE_AA)
            pyautogui.keyDown('a')
            
            pyautogui.keyUp('a')
        elif(tb1>150 and rt1<90 and bl1 < 90):
            
            cv2.putText(frame,"Right",(400,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2,cv2.LINE_AA)
            pyautogui.keyDown('d')
            pyautogui.keyUp('d')
        else:
            
            cv2.putText(frame,"Forward",(400,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2,cv2.LINE_AA)
            pyautogui.keyDown('j')
            pyautogui.keyUp('j')

        cv2.circle(img1,t1,1,(0,0,255),5)
        cv2.circle(img1,b1,1,(0,0,255),5)
        cv2.circle(img1,l1,1,(0,0,255),5)
        cv2.circle(img1,r1,1,(0,0,255),5)
        
    
    
    Control2
    img2 = frame[10:300,0:300]

    hsv2 = cv2.cvtColor(img2,cv2.COLOR_BGR2HSV)
    
    skin2 = cv2.inRange(hsv2,l_r,u_r)
   
    skin2 = cv2.erode(skin2,kernel,iterations=2)
    skin2 = cv2.dilate(skin2,kernel,iterations=2)

    skin2 = cv2.medianBlur(skin2,11)
    cont2,hier2 = cv2.findContours(skin2,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    if(len(cont2)!=0):
        cont2 = max(cont2,key=lambda x: cv2.contourArea(x))
        cv2.drawContours(img2,[cont2],-1,(255,255,0),2)

        hull2 = cv2.convexHull(cont2)
        cv2.drawContours(img2,[hull2],-1,(0,255,255),2)


        t2=tuple(hull2[hull2[:,:,1].argmin()][0])
        b2=tuple(hull2[hull2[:,:,1].argmax()][0])
        l2=tuple(hull2[hull2[:,:,0].argmin()][0])
        r2=tuple(hull2[hull2[:,:,0].argmax()][0])
        
        tb2 = np.linalg.norm(t2[-1]-b2[-1])
        if(tb2<150):
            cv2.putText(frame,"Down",(10,250),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2,cv2.LINE_AA)
            pyautogui.keyDown('s')
            pyautogui.keyUp('s')
        elif(tb2>150):
            cv2.putText(frame,"Up",(10,250),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2,cv2.LINE_AA)
            pyautogui.keyDown('w')
            pyautogui.keyUp('w')
            
            

        cv2.circle(img2,t2,1,(0,0,0),5)
        cv2.circle(img2,b2,1,(0,0,0),5)
        cv2.circle(img2,l2,1,(0,0,0),5)
        cv2.circle(img2,r2,1,(0,0,0),5)
    
    
    
    
    cv2.imshow("Video",frame)
    cv2.imshow("Hand 2",img1)
    cv2.imshow("Hand1", img2)
    
    if cv2.waitKey(1) & 0xFF==ord('q'):  
        
        cv2.destroyAllWindows()
        break
```
